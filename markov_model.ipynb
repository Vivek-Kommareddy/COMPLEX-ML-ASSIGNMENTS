**Exercise 5-1:**  **Observable Markov Model**

Consider a simple 3-state Markov model of the weather. We assume that once a day (e.g. at noon), the weather is observed as being one of the following: <br>

- State 1: rain or snow <br>
- State 2: cloudy <br>
- State 3: sunny <br>

We postulate that the weather on day t is characterized by a single one of the three states above. We have the following transition probabilities:

a. Given that the weather on day 1 (t=1) is rainy, what is the probability that the weather on the
next five days is: {S1, S2, S3, S3, S1}?

>Answer:

```python
import numpy as np
```

To compute the probability of a sequence of weather states, we need to multiply the transition probabilities for each state transition in the sequence.

Probability of the sequence {S1, S2, S3, S3, S1}:

$ P = P(s1\rightarrow s2) \times P(s2\rightarrow s3) \times P(s3\rightarrow s3) \times P(s3\rightarrow s1) $

$ P = 0.3 \times 0.2 \times 0.8 \times 0.1 $

$ P = 0.0048 $



```python
# Transition matrix
transition_matrix = np.array([
    [0.4, 0.3, 0.3],
    [0.2, 0.6, 0.2],
    [0.1, 0.1, 0.8]
])

# Initial state (rainy)
initial_state = np.array([1, 0, 0])

sequence = [0, 1, 2, 2, 0]  # S1 corresponds to 0, S2 corresponds to 1, S3 corresponds to 2

prob_sequence = 1.0
for i in range(len(sequence) - 1):
    prob_sequence *= transition_matrix[sequence[i], sequence[i+1]]

print(f"Probability of sequence {{S1, S2, S3, S3, S1}}: {prob_sequence:.4f}")
```

b. Given that the model is in a known state, what is the probability it stays in that state for exactly d days?

> Answer:


Let's denote the probability of staying in a state i for exactly d days as $ P_i(d) $.

For state 1 (rain or snow):
$ P_1(d) = 0.4^{d-1} $

For state 2 (cloudy):
$ P_2(d) = 0.6^{d-1} $

For state 3 (sunny):
$ P_3(d) = 0.8^{d-1} $



```python

def prob_stay_in_state(d, transition_matrix):
    stay_probabilities = np.diag(transition_matrix)  # Diagonal values are stay probabilities

    # Probability of staying in the same state for d-1 transitions
    stay_prob_d = stay_probabilities ** (d-1)

    print(f"Probability of staying in the same state for d={d} days:")
    print(f"   State 1 (S1): {stay_prob_d[0]:.4f}")
    print(f"   State 2 (S2): {stay_prob_d[1]:.4f}")
    print(f"   State 3 (S3): {stay_prob_d[2]:.4f}")

# d is the number of days
prob_stay_in_state(1, transition_matrix)
prob_stay_in_state(2, transition_matrix)
prob_stay_in_state(3, transition_matrix)
```

c. What is the expected number of observations in a state? Note:  $$\sum_{d=1}^\infty da^{d-1} = \frac{1}{(a-1)^2}$$ when |a| < 1

> Answer:


The expected number of observations in a state i is given by the formula:

$ \sum_{d=1}^\infty d \times P_i(d) $

Using the given formula:
$ \sum_{d=1}^\infty da^{d-1} = \frac{1}{(a-1)^2} $

For state 1 (rain or snow):
$ \sum_{d=1}^\infty d \times 0.4^{d-1} = \frac{1}{(0.4-1)^2} $
$ = \frac{1}{(-0.6)^2} = \frac{1}{0.36} = 2.7778 $

For state 2 (cloudy):
$ \sum_{d=1}^\infty d \times 0.6^{d-1} = \frac{1}{(0.6-1)^2} $
$ = \frac{1}{(-0.4)^2} = \frac{1}{0.16} = 6.25 $

For state 3 (sunny):
$ \sum_{d=1}^\infty d \times 0.8^{d-1} = \frac{1}{(0.8-1)^2} $
$ = \frac{1}{(-0.2)^2} = \frac{1}{0.04} = 25 $

So, the expected number of observations in state 1 is approximately 2.7778, in state 2 is approximately 6.25, and in state 3 is approximately 25.

```python

def expected_observations(transition_matrix):
    stay_probabilities = np.diag(transition_matrix)
    expected_observations = 1 / (1 - stay_probabilities)**2

    return expected_observations

expected_obs = expected_observations(transition_matrix)
print(f"c. Expected number of observations in a state:")
print(f"   State 1 (S1): {expected_obs[0]:.4f}")
print(f"   State 2 (S2): {expected_obs[1]:.4f}")
print(f"   State 3 (S3): {expected_obs[2]:.4f}")
```

**Exercise 5-2:** **Hidden Markov Model** <br>

Suppose we want to determine the average annual temperature at a particular location on the earth over a series of years. To make it interesting, suppose the years we are concerned with lie in the distant past, before thermometers were invented.  Since we can't go back in time, we instead look for indirect evidence of the temperature.

To simplify the problem, we only consider two annual temperatures, "hot" and "cold". Suppose that modern evidence indicates that the probability of a hot year followed by another hot year is 0.7 and the probability that a cold year is followed by another cold year is 0.6.  We'll assume that these probabilities held in the distant past as well.  The information so far can be summarized as:

$$
\begin{bmatrix}
& H & C \\
H & 0.7 & 0.3 \\
C & 0.4 & 0.6 \\
\end{bmatrix}
$$

where H is "hot" and C is "cold'.

Also suppose that current research indicates a correlation between the size of tree growth rings and temperature.  For simplicity, we only consider three different tree ring sizes, small, medium, and large or S, M, and L, respectively.  Finally, suppose that based on available evidence, the probabilistic relationship between annual temperature and tree ring sizes is given by

$$
\begin{array}{c|ccc}
& S & M & L \\
\hline
H & 0.1 & 0.4 & 0.5 \\
C & 0.7 & 0.2 & 0.1 \\
\end{array}
$$

Suppose the initial distribution is $\pi$ = (0.6, 0.4).  You have the following observation sequence {S, L, L}. Compute the distribution of states as well as the most likely sequence.

```python
import numpy as np

# Define transition probabilities
transition_probs = np.array([[0.7, 0.3],
                             [0.4, 0.6]])

# Define emission probabilities
emission_probs = np.array([[0.1, 0.4, 0.5],
                           [0.7, 0.2, 0.1]])

# Initial distribution
initial_dist = np.array([0.6, 0.4])

# Observation sequence
observations = ['S', 'L', 'L']

# Forward Algorithm
def forward_algorithm(obs_seq, init_dist, trans_probs, emit_probs):
    num_states = len(init_dist)
    num_obs = len(obs_seq)

    alpha = np.zeros((num_obs, num_states))
    alpha[0] = init_dist * emit_probs[:, observations.index(obs_seq[0])]

    for t in range(1, num_obs):
        for j in range(num_states):
            alpha[t, j] = np.sum(alpha[t-1] * trans_probs[:, j]) * emit_probs[j, observations.index(obs_seq[t])]

    return alpha

# Backward Algorithm
def backward_algorithm(obs_seq, trans_probs, emit_probs):
    num_states = trans_probs.shape[0]
    num_obs = len(obs_seq)

    beta = np.zeros((num_obs, num_states))
    beta[-1] = 1

    for t in range(num_obs - 2, -1, -1):
        for i in range(num_states):
            beta[t, i] = np.sum(beta[t+1] * trans_probs[i, :] * emit_probs[:, observations.index(obs_seq[t+1])])

    return beta

# Calculate state distribution and most likely sequence
def calculate_distribution_and_sequence(alpha, beta):
    state_distribution = alpha * beta
    state_distribution /= np.sum(state_distribution, axis=1)[:, np.newaxis]

    most_likely_sequence = np.argmax(state_distribution, axis=1)

    return state_distribution, most_likely_sequence

# Main
alpha = forward_algorithm(observations, initial_dist, transition_probs, emission_probs)
beta = backward_algorithm(observations, transition_probs, emission_probs)
state_distribution, most_likely_sequence = calculate_distribution_and_sequence(alpha, beta)

print("State Distribution:")
print("\n".join([f"Day {i+1}: {state_distribution[i]}" for i in range(len(observations))]))
print("\nMost Likely Sequence:")
print(most_likely_sequence)
print(" -> ".join([observations[most_likely_sequence[i]] for i in range(len(most_likely_sequence))]))
```
